```python
# 05
# if
# --- Guess a random number ---
import random
secret_number = random.randint(1, 100)
correct = False

while correct == False:
  print('\nGuess the number')
  user_number = int(input('Enter a number: '))
  if secret_number == user_number:
    correct = True
    print('You guessed it!\n')
  elif secret_number > user_number:
    print('The number is higher\n')
  elif secret_number < user_number:
    print('The number is lower\n')
```

```python
# 06
# if
# --- IMC calculator ---
print('IMC Calculator\n')
weight = float(input('Enter your weight (kg): '))
height = float(input('Enter your height (m): '))
imc = weight / (height * height)

if imc <= 18.5:
  print('Under weight')
elif 18.5 <= imc < 25:
  print('Healthy weight')
elif 25 <= imc < 30:
  print('Over weight')
elif imc >= 30:
  print('Obesity')
```

```python
# 10
'''
Prompt:
**Problem: Calculate the factorial of a number**
Write a program that prompts the user to enter an integer and then calculates the factorial of that number using a for loop.
The factorial of an integer n is defined as the product of all positive integers from 1 to n.
For example, if the user enters 5, the program should calculate 5 factorial (5!) which is equal to 5 x 4 x 3 x 2 x 1 = 120.
'''

num_factorial = int(input('Give me a number: '))
list_factorial = list(range(1,num_factorial))

total_factorial = num_factorial
for element in list_factorial:
  total_factorial = total_factorial * element

print('Total is ', total_factorial)
```

```python
# 11
'''
Prompt:
**Problem: Counting Vowels in a String**
Write a program that prompts the user to enter a text string and then counts the number of vowels (a, e, i, o, u) in that string.
The program should display the total count of each vowel individually.
For example, if the user enters the string "Hola, cómo estás?", the program should count and display the number of "a," "e," "i," "o," and "u" in the string.
'''
a_sum = 0
e_sum = 0
i_sum = 0
o_sum = 0
u_sum = 0

user_text = input('Give me a sentence: ')
for vowel in user_text:
  if vowel == 'a':
    a_sum = a_sum + 1
  elif vowel == 'e':
    e_sum = e_sum + 1
  elif vowel == 'i':
    i_sum = i_sum + 1
  elif vowel == 'o':
    o_sum = o_sum + 1
  elif vowel == 'u':
    u_sum = u_sum + 1

print(f'Number of a in the sentence is... {a_sum}')
print(f'Number of e in the sentence is... {e_sum}')
print(f'Number of i in the sentence is... {i_sum}')
print(f'Number of o in the sentence is... {o_sum}')
print(f'Number of u in the sentence is... {u_sum}')
```

```python
# 12

'''
Prompt:
** Problem: Calculate the Average of a List of Numbers **
Write a program that calculates the average of a list of numbers entered by the user.
The program should prompt the user to enter how many numbers they want to average and then input those numbers.
Then, it calculates the average of the entered numbers.
For example, if the user enters that they want to average 5 numbers (e.g., 10, 15, 20, 25, 30), the program should calculate the average, which in this case is (10 + 15 + 20 + 25 + 30) / 5 = 20.
'''

number_list = []
usr_mean_opc = '-'

while usr_mean_opc != '*':
  number_mean = int(input('Write a number: '))
  number_list.append(number_mean)

  usr_mean_opc = input('To stop writting numbers press * if you want to keep writting write another character')

  if usr_mean_opc == '*':
    continue

print('The mean is: ', sum(number_list)/len(number_list))

```
